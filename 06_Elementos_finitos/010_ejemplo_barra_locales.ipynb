# Ejemplo 3

$L = 40 \ \text{cm}$, $A_{1} = 25 \ \text{cm}^{2}$, $A_{2} = 16 \ \text{cm}^{2}$ y $E = 2000 \ \text{Kg}/\text{cm}^{2}$

\begin{equation}
    A(x) = 25 - \frac{9}{40} x
\end{equation}

<img align="left" src="imagenes/012_ejemplo_barra_1.svg">

### Un elemento de dos nodos

<img align="left" src="imagenes/012_ejemplo_barra_2.svg">

\begin{equation}
    \int_{0}^{L} B^{T} D \ B \ dx \ u = \int_{0}^{L} q \ N^{T} dx + f
\end{equation}

desplazamientos

\begin{equation}
    N =
    \begin{bmatrix}
        1 - \frac{1}{40} x & \frac{1}{40} x
    \end{bmatrix}
\end{equation}

deformaciones

\begin{equation}
    B = \frac{d N}{d x} =
    \begin{bmatrix}
        -\frac{1}{40} & \frac{1}{40}
    \end{bmatrix}
\end{equation}

matriz constitutiva

\begin{equation}
    D = E A = 50000 - 450 x
\end{equation}

reemplazando

\begin{equation}
    \int_{0}^{40}
    \begin{bmatrix}
        -\frac{1}{40} \\
        \frac{1}{40}
    \end{bmatrix}
    \begin{bmatrix}
        50000 - 450 x
    \end{bmatrix}
    \begin{bmatrix}
        -\frac{1}{40} & \frac{1}{40}
    \end{bmatrix} dx
    \begin{bmatrix}
    0 \\
    u_{2}
    \end{bmatrix} =
    \begin{bmatrix}
    F_{1} \\
    -300
    \end{bmatrix}
\end{equation}

integrando y sumando

\begin{equation}
    \begin{bmatrix}
        1025 & -1025 \\
        -1025 & 1025
    \end{bmatrix}
    \begin{bmatrix}
    0 \\
    u_{2}
    \end{bmatrix} =
    \begin{bmatrix}
    F_{1} \\
    -300
    \end{bmatrix}
\end{equation}

resolviendo

\begin{align}
    F_{1} &= 300 \ \text{Kg} \\
    u_{2} &= -0.293 \ \text{cm}
\end{align}

###Desplazamientos, deformaciones y esfuerzos

desplazamientos

\begin{equation}
    u = N u =
    \begin{bmatrix}
        1 - \frac{1}{40} x & \frac{1}{40} x
    \end{bmatrix}
    \begin{bmatrix}
    0 \\
    -0.293
    \end{bmatrix} = -0.007325 x
\end{equation}

deformación normal

\begin{equation}
    \varepsilon = B u =
    \begin{bmatrix}
        -\frac{1}{40} & \frac{1}{40}
    \end{bmatrix}
    \begin{bmatrix}
    0 \\
    -0.293
    \end{bmatrix} = -0.007325
\end{equation}

esfuerzo normal

\begin{equation}
    \sigma = E \varepsilon = -14.65 \ \text{Kg}/\text{cm}^{2}
\end{equation}
